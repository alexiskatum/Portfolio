```python
from sklearn import preprocessing
```

```python
# drop les lignes que la valeur 'price' car il est tres influent
df = df.dropna(subset=['price'])

#median (assez influent)
#      df['price'].fillna(value=np.round(df['price'].median(), decimals=2), inplace=True)
df['normalized-losses'].fillna(value=np.round(df['normalized-losses'].mean(), decimals=2), inplace=True)
df['bore'].fillna(value=np.round(df['bore'].mean(), decimals=2), inplace=True)
df['horsepower'].fillna(value=np.round(df['horsepower'].median(), decimals=2), inplace=True)

#mean
df['num-of-doors'].fillna(value=np.round(df['num-of-doors'].mean(), decimals=2), inplace=True)
df['stroke'].fillna(value=np.round(df['stroke'].mean(), decimals=2), inplace=True)
df['peak-rpm'].fillna(value=np.round(df['peak-rpm'].mean(), decimals=2), inplace=True)

df
```
