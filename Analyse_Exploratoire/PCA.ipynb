```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from numpy import linalg as LA
from sklearn.decomposition import PCA
from fanalysis.pca import PCA as FanalysisPCA

df = pd.read_excel('DataLiqueurs1.xlsx', index_col=0)

display(df)

print(df.dtypes)
print('\n')
print(df['Region'].value_counts())
print('\n')
print(df['Soil'].value_counts())
print('\n')
```

#### Les données contiennent 21 lignes (individus) et 8 colonnes (variables): Les deux premières colonnes sont des variables qualitatives (catégorielles) region et soil. Les colonnes restantes sont quantitatives (variables continues).

### separation des variables qualitative et quantitatives 

```python
df2 = df.iloc[:, 0:2]
df1 = df.iloc[:, 2:14]

display(df2)
```

#### La variable Region a trois modalités (Saumur, Bourgueil ou Chinon) et la variable soil a 4 modalités (Reference, Env1, Env2 or Env4).

```python
display(df1)
```

### correlation

```python
df1.corr().style.background_gradient(cmap='coolwarm')
```

### Analyse en ACP

```python
scaler = preprocessing.StandardScaler()
sdf1= scaler.fit_transform(df1)
corr_matrix = np.corrcoef(sdf1.transpose())
val, vec = np.linalg.eig(corr_matrix )

print("La matrice de corrélation: \n", corr_matrix.round(2), '\n')
print('Les valeurs propres sont:', '\n', val.round(3), '\n')
print('Les vecteurs propres sont:', '\n', vec.round(3))
```

#### Selon la règle de Kaiser on doit retenir les deux composantes principales lambda > 1, donc nous retiendront les valeur propres suivants : [ 6.829 et 2.219 ]

### Réalisation ACP

```python
acp = PCA(svd_solver='full', n_components=None)
acp2 = PCA(svd_solver='full', n_components=2) # 2 composant

adf1=acp.fit_transform(sdf1)
adf=acp2.fit_transform(sdf1) # 2 composant

print(adf1)
print(adf) # 2 composan
```

### Représentation des individus

```python
cont = adf1**2
n = df1.shape[0]
for j in range(df1.shape[1]):
    cont[:,j] = cont[:,j]/(n*val[j])
    
display(pd.DataFrame({'CTR_1':cont[:,0],'CTR_2':cont[:,1]}, index=df1.index))


fig, axes = plt.subplots(figsize=(15,10))
axes.set_xlim(-5,8)
axes.set_ylim(-5,5)

for i in range(df1.shape[0]):
    plt.annotate(df1.index[i],(adf[i,0],adf[i,1]))
plt.plot([-5,8],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-5,5],color='silver',linestyle='-',linewidth=1)
plt.ylabel('Axe factoriel 2 ')
plt.xlabel('Axe factoriel 1 ')
plt.show()
```

#### Les individus 1VAU, 2ING sont déterminants pour le premier axe factoriel.
#### Les individus T1 et T2 sont déterminants du deuxième axe factoriel.
#### Les individus proches du centre sont les moins déterminants.

# AFDM

### Codage des données qualitatives 

```python
dfm = pd.get_dummies(df2)
display(dfm)
```

### Proportion et Standardisation des indicatrices 

```python
p_k = np.mean(dfm.values,axis=0)
ZD = dfm.values/np.sqrt(p_k)

print(p_k, '\n')
display(ZD.round(3))
```

### Fusion des donnees qualitative et quantitative

```python
Z = np.concatenate((sdf1, ZD),axis=1)
display(Z.round(3))

# Dimension de la matrice de données préparées
k = Z.shape[1]
print('\n')
print('\n')
print('Nbr variable : ', k)

col = list(df1.columns)+list(dfm.columns)
print('\n')
print(col)
```

### Analyse AFDM

```python
#instanciation - les données sont déjà standardisées
afdm = FanalysisPCA(std_unit=False,row_labels=df.index,col_labels=col)
#lancement des calculs
afdm.fit(Z)
None
```

```python
fix,ax = plt.subplots(figsize=(12,8))
ax.axis([-6,+8,-6,+8])
ax.plot([-6,+8],[0,0],color='silver',linestyle='--')
ax.plot([0,0],[-6,+8],color='silver',linestyle='--')
ax.set_xlabel("Dim.1")
ax.set_ylabel("Dim.2")
plt.title("Représentation des individus")
for i in range(df2.shape[0]):
    ax.text(afdm.row_coord_[i,0],afdm.row_coord_[i,1],df2.index[i],color='firebrick') 
plt.show()
```

### Comparaison des résultats

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_subplot(1, 2, 1)
#positionnement des individus dans le premier plan
ax1.set_xlim(-10,10)
ax1.set_ylim(-10,10)
plt.title("Représentation des individus1")
plt.scatter(adf[:,0],adf[:,1],s=5,c='firebrick')
for i in range(df1.shape[0]) :
    ax1.annotate(df1.index[i],(adf[i,0],adf[i,1]+.3),color='#1f77b4')
#ajouter les axes
ax1.plot([-10,10],[0,0],color='silver',linestyle='-',linewidth=1)
ax1.plot([0,0],[-10,10],color='silver',linestyle='-',linewidth=1)
ax1.set(xlabel='Axe factoriel 1 (74%)', ylabel='Axe factoriel 2 (14%)')

ax2 = fig.add_subplot(1, 2, 2)
#positionnement des individus dans le premier plan
ax2.set_xlim(-10,10)
ax2.set_ylim(-10,10)
plt.title("Représentation des individus2")
plt.scatter(afdm.row_coord_[:,0],afdm.row_coord_[:,1],s=5,c='firebrick')
for i in range(df2.shape[0]) :
    ax2.annotate(df2.index[i],(afdm.row_coord_[i,0],afdm.row_coord_[i,1]+.3),color='#1f77b4')
#ajouter les axes
ax2.plot([-10,10],[0,0],color='silver',linestyle='-',linewidth=1)
ax2.plot([0,0],[-10,10],color='silver',linestyle='-',linewidth=1)
ax2.set(xlabel='Axe factoriel 1', ylabel='Axe factoriel 2')
plt.show()
```

#### L'ajout de varaibles qualitatives n'a pas eu d'impact sur les varaibles ayant une forte détermination des axes factoriels (1VAU et 2ING) et (T1 et T2). La similarité a été préservée
#### Par contre pour les variables situées au centre, cet ajout a eu un impact. On voit apparaitre deux a trois regroupements d'individus
