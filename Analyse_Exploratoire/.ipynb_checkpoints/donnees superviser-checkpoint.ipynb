# Import Library and DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('ListeEmployes.xlsx')
display(data)
```

# Analyse Data

```python
print('The dataframe contains', data.shape[0], 'individuals and', data.shape[1], 'variables.', '\n')
print('The variables', data.columns[0], '(', data.dtypes[0], ')', data.columns[1], '(', data.dtypes[1], ')', 
      data.columns[4], '(', data.dtypes[4], ')', 'are qualitative.', '\n', 'The variables', data.columns[2], '(', data.dtypes[2], ')', 
      data.columns[3], '(', data.dtypes[3], ')', 'are both quantitative, while salary is discrete and bonus is continuous.', '\n')
print('The variable Equipe contains the following 9 categories:', '\n', data['Equipe'].unique(), '\n', '\n',
      'And here is the count for each category of the variable Team:', '\n', data['Equipe'].value_counts())
```

```python
diag = data.Equipe.value_counts()
diag.plot.pie(figsize = (4, 4))
```

```python
salary = data['Salaire']

print('Here are 5 characteristics of central tendencies for the variable Salary:', '\n', '\n', 'The mean is:', salary.mean())
print('The median is:', salary.median(), '\n')

print('The first quartile is:', salary.quantile(0.25))
print('The second quartile is:', salary.quantile(0.5))
print('The third quartile is:', salary.quantile(0.75))
```

```python
salary.hist(figsize = (5,3))
```

```python
print('Since the skewness coefficient,', salary.skew(axis = 0, skipna = True), ''', is greater than zero, 
      the distribution of the variable should be shifted to the left, as we can observe on the above histogram. 
      Furthermore, this same coefficient indicates that the mean is supposedly greater than the median, 
      as confirm when we go back at the 5 characteristics of the variable Salary.''')
```
