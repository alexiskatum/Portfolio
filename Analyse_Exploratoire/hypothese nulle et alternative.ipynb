# Hypothese Nulle et Alternative

#### Notre Hypothese nulle (H0) sera que la moyenne de prix des vehicule EST EGALE a 20 000 et l' Hypothese Aleternative (HA) sera que cette meme moyenne N'EST PAS EGALE a 20 000

```python
ho = stats.ttest_1samp(df2010['price'], 20000, alternative = 'two-sided')
print( 'Chi2 : ', ho[0] )
print('pvalue : ', ho[1])
print(ho)
```

#### Puisque le PVALUE est plus petit que 0.1, 0.05 et 0.01, l'hypothese nulle sera rejeter que l'intervale de confiance soit 90%, 95% ou bien 99%

# Interval visee

```python
print('intervale de confiance 90%', stats.t.interval(0.9, len(df2010['price'])-1, loc=np.mean(df2010['price']), scale=stats.sem(df2010['price'])))
print('intervale de confiance 95%', stats.t.interval(0.95, len(df2010['price'])-1, loc=np.mean(df2010['price']), scale=stats.sem(df2010['price'])))
print('intervale de confiance 99%', stats.t.interval(0.99, len(df2010['price'])-1, loc=np.mean(df2010['price']), scale=stats.sem(df2010['price'])))
```

#### Nous choisiront l'Hypothese selon le pourcentage de H0 que nous voulons rejeter ex: ~ 2% on irez entre ~[18 642 et 18 792] ou ~[19 746 et 19 896] prenons 18 700

```python
ho = stats.ttest_1samp(df2010['price'], 18700, alternative = 'two-sided')
print( 'Chi2 : ', ho[0] )
print('pvalue : ', ho[1])
print(ho)
```

#### Puisque le PVALUE est plus petit que 0.1 et 0.05, l'hypothese nulle sera rejeter que l'intervale de confiance soit 90% ou 95%. Serte elle ne sera pas rejeter si l'intervale de confiance est de 99% car PVALUE (~ 0.02) est plus grang que 0.01
