```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn import preprocessing
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from datatile.summary.df import DataFrameSummary
from sklearn import tree
from joblib import dump, load
```

```python
cols_name = ['Age','Workclass','fnlwgt','Education','Education-num','Marital-Status', 'Occupation', 'Relationship', 'Race', 
             'Sex', 'Capital-gain', 'Capital-loss', 'Hours-per-week', 'Native_country', 'Target']
df = pd.read_csv('adult.data', names=cols_name, index_col=False)

observation = df.shape[0]
variables= df.shape[1]

display(df.head())
print(df.info())
print('\n')
print(f'Il y a {observation} individus (lignes) et {variables} variables (colonnes) dans cet ensemble de données') 
```

#### Il y a 6 variables quantitatives et 9 variables qualitatives dont la cible qui est Target

### Supprimer les individus ayant des valeurs manquantes 

```python
rows_with_unknowns=df.apply(lambda x: (x==' ?').sum(), axis=1)>0
index=df[rows_with_unknowns].index.tolist()
df=df.drop(index, axis=0)

for i in df.iloc[:,: 15]:
    print(f'================== Variable {i} ================== \n')
    print(f'Le nombre de modalités pour la variable {i} est: ', df.groupby(i).ngroups)
    print(f'La fréquence des modalités de la variable {i} est: \n',df[i].value_counts())
```

```python
observation = df.shape[0]
variables= df.shape[1]
print(f'Il y a {observation} individus (lignes) et {variables} variables (colonnes) dans cet ensemble de données')
```

## Valeurs Aberrantes 

```python
data=df

plt.rcParams.update({'font.size': 14})
fig = plt.figure(figsize=(15, 15))

plt.subplot(231)
plt.boxplot(data['Age'])
plt.xlabel("Age", fontsize=16)

plt.subplot(232)
plt.boxplot(data['fnlwgt'])
plt.xlabel("fnlwgt", fontsize=16)

plt.subplot(233)
plt.boxplot(data['Education-num'])
plt.xlabel("Education-num", fontsize=16)

plt.subplot(234)
plt.boxplot(data['Capital-gain'])
plt.xlabel("Capital-gain", fontsize=16)

plt.subplot(235)
plt.boxplot(data['Capital-loss'])
plt.xlabel("Capital-loss", fontsize=16)

plt.subplot(236)
plt.boxplot(data['Hours-per-week'])
plt.xlabel("Hours-per-week", fontsize=16)
```

### Supprimer les individus ayant des valeurs aberrantes

```python
mu= np.mean(data['Age'] )
sigma = np.std (data['Age'])
data=data[np.abs(data['Age'] - mu) <= (2 * sigma)]

mu= np.mean(data['fnlwgt'] )
sigma = np.std (data['fnlwgt'])
data=data[np.abs(data['fnlwgt'] - mu) <= (2 * sigma)]

mu= np.mean(data['Education-num'] )
sigma = np.std (data['Education-num'])
data=data[np.abs(data['Education-num'] - mu) <= (2 * sigma)]

plt.rcParams.update({'font.size': 14})
fig = plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.boxplot(data['Age'])
plt.xlabel("Age", fontsize=16)

plt.subplot(132)
plt.boxplot(data['fnlwgt'])
plt.xlabel("fnlwgt", fontsize=16)

plt.subplot(133)
plt.boxplot(data['Education-num'])
plt.xlabel("Education-num", fontsize=16)
```

```python
observation = data.shape[0]
variables= data.shape[1]

df=data
df.reset_index(drop=True, inplace=True)

print(f'Il y a {observation} individus (lignes) et {variables} variables (colonnes) dans cet ensemble de données')
```

## Encodage
### Target

```python
index=0
for j in df['Target']:
    X=df.loc[index, 'Target']
    if X==' <=50K':
        df.loc[index, 'Target']=0
    else:
        df.loc[index, 'Target']=1
    index+=1

df["Target"] = df["Target"].astype(str).astype(int)
df['Target'].head()
```

### Variables Qualitatives

```python
print(df.info())
```

```python
for i in df:
    if df[i].dtypes == 'object':
        df[i] = LabelEncoder().fit_transform(df[i])

display(df)
DataFrameSummary(df).summary()
```

## Correlation

```python
sns.pairplot(df)
```

```python
features_mean= list(df.columns[0:15])
plt.figure(figsize=(15,15))
sns.heatmap(df[features_mean].corr().round(2), annot=True, square=True, cmap='coolwarm')

plt.title('Matrice de corrélation', fontsize = 18, color = 'black')
```

```python
featcols= ['Age','Education-num','Marital-Status', 'Relationship', 'Sex','Capital-gain', 'Hours-per-week']
feat=df['Age'], df['Education-num'],df['Marital-Status'], df['Relationship'], df['Sex'], df['Capital-gain'], df['Hours-per-week']
Features = pd.DataFrame(np.c_[feat], columns= featcols)
cible = df['Target']

scaler = StandardScaler()
scaler.fit(Features)
scaled = scaler.fit_transform(Features).round(5)

scaled_df = pd.DataFrame(scaled, columns=Features.columns)
display(scaled_df)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(scaled, cible, test_size=0.20, random_state= 44)

print(f"================== Pour les données d'entrainement ================== ")
print(f'Il y a {X_train.shape[0]} individus (lignes) et {X_train.shape[1]} variables (colonnes) dans cet ensemble de données')
print(f'Il y a {Y_train.shape[0]} individus (lignes) pour la cible. \n')
print(f"================== Pour les données de test ================== ")
print(f'Il y a {X_test.shape[0]} individus (lignes) et {X_test.shape[1]} variables (colonnes) dans cet ensemble de données')
print(f'Il y a {Y_test.shape[0]} individus (lignes) pour la cible.')
```

## Model d'apprentissage

### Regression 

```python
LogisticR=LogisticRegression()
LogisticR.fit(X_train, Y_train)
```

```python
z=LogisticR.predict(X_train)
accuracy1=LogisticR.score(X_train,Y_train)
print(z.shape)
print("Train Accuracy : ", accuracy1)

print('\n')

z=LogisticR.predict(X_test)
accuracy_pre1=LogisticR.score(X_test,Y_test)
print(z.shape)
print("Test Accuracy : ", accuracy_pre1)
```

### KNN

```python
classifier_knn = KNeighborsClassifier(n_neighbors = 18, metric = 'minkowski', p = 2)
classifier_knn.fit(X_train, Y_train)
```

```python
z=classifier_knn.predict(X_train)
accuracy2=classifier_knn.score(X_train,Y_train)
print(z.shape)
print("Train Accuracy : ", accuracy2)

print('\n')

z=classifier_knn.predict(X_test)
accuracy_pre2=classifier_knn.score(X_test,Y_test)
print(z.shape)
print("Test Accuracy : ", accuracy_pre2)
```

```python
k_range=range(1,26)
scores=[]
for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred=knn.predict(X_test)
    scores.append(metrics.accuracy_score(Y_test,Y_pred))
print(scores)

plt.plot(k_range, scores)
plt.xlabel("valeur de k pour KNN")
plt.ylabel("Testing accuracy")
```

### Arbre de decision

```python
classifier_dt = DecisionTreeClassifier(criterion = 'entropy', random_state=35, max_depth=8, min_samples_leaf=2)
classifier_dt.fit(X_train, Y_train)
```

```python
z=classifier_dt.predict(X_train)
accuracy3=classifier_dt.score(X_train,Y_train)
print(z.shape)
print("Accuracy : ", accuracy3)

print('\n')

z=classifier_dt.predict(X_test)
accuracy_pre3=classifier_dt.score(X_test,Y_test)
print(z.shape)
print("Accuracy : ", accuracy_pre3)
```

```python
tree.plot_tree(classifier_dt.fit(X_train, Y_train))
```

```python
print("La précision du modèle de Regression logistique est: {0:7.7f} " .format(accuracy_pre1))
print("La précision du modèle de KNN est: {0:7.7f} " .format(accuracy_pre2))
print("La précision du modèle de Decision Tree est: {0:7.7f} ".format(accuracy_pre3))
```

```python
dump(classifier_dt.fit(X_train, Y_train), 'meilleur_modeleDT')
```
